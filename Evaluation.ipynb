```python
# Importing everything

from sklearn import tree
from sklearn import metrics
import numpy as np
from matplotlib import pyplot as plt
import csv
import numpy as np
from utils import load_from_csv ## utils file in moodle

# load from dataset
train_examples, train_labels, features, prediction = load_from_csv("./compass.csv")

# Create partition
trainingPartition = round(train_labels.size*0.8)

trainingData = train_examples[:trainingPartition]
testingData = train_examples[trainingPartition:]

trainingLabel = train_labels[:trainingPartition]
testingLabel = train_labels[trainingPartition:]
```

A decision tree configuration is a set of parameters that one can use to build decision trees. Propose 6 configurations that are likely to provide different topologies and caracteristics

```python
clf_low_depth = tree.DecisionTreeClassifier(max_depth=2)
clf_low_leaf = tree.DecisionTreeClassifier(min_samples_leaf=2)
clf_medium_leaf = tree.DecisionTreeClassifier(min_samples_leaf=10)
clf_random_splitter = tree.DecisionTreeClassifier(splitter='random')
# The default decision tree has best split strategy, no max depth and 1 sample leaf minimum
clf_default = tree.DecisionTreeClassifier()
trees = [clf_low_depth,clf_low_leaf,clf_medium_leaf,clf_random_splitter,clf_default]
```

Train a decision tree for each of the previous configurations on the full dataset

```python
def fit(trainingData,trainingLabel):
    return lambda x : x.fit(trainingData,trainingLabel)
def predict(testingData):
    return lambda x : x.predict(testingData)

fit1 = fit(trainingData,trainingLabel)
predict1 = predict(testingData)
fitted = list(map(fit1,trees))
predictions = list(map(predict1,fitted))


tree.plot_tree(fitted[0])
```

Propose an evaluation in terms of training and testing accuracies using $5$-cross validation on two decision trees that have different typologies

```python
from sklearn.model_selection import KFold
# Create 5-cross validation splitter
cv_5 = KFold(n_splits=5)

def createSets(train_index,test_index,train_examples,train_labels):
    #construct Training and Testing data sets
    trainExamples = [elem for index,elem in enumerate(train_examples) if (index in train_index)]
    trainLabels = [elem for index,elem in enumerate(train_labels) if (index in train_index)]
    testExamples = [elem for index,elem in enumerate(train_examples) if (index in test_index)]
    testLabels = [elem for index,elem in enumerate(train_labels) if (index in test_index)]
    return trainExamples,trainLabels,testExamples,testLabels
    
toTrain = [trees[1],trees[3]]
fittedCross = []
crossScore = [[],[]]
for train_index, test_index in cv_5.split(train_examples):
    trainExamples,trainLabels,testExamples,testLabels = createSets(train_index,test_index,train_examples,train_labels)
    # train it for the index and test it in each iteration
    newFittedTrees = list(map(fit(trainExamples,trainLabels),toTrain))
    fittedCross.append(newFittedTrees)

    predictions = list(map(predict(testExamples),newFittedTrees))
    accuracy = [metrics.accuracy_score(pred,testLabels) for pred in predictions]
    crossScore[0].append(accuracy[0])
    crossScore[1].append(accuracy[1])
    print(accuracy)
    print(crossScore)

# Plot both predictions accuracy for each cross Test??
```
