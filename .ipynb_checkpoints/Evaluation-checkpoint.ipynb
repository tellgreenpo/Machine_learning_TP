A decision tree configuration is a set of parameters that one can use to build decision trees. Propose 6 configurations that are likely to provide different topologies and caracteristics

```python
clf_low_depth = tree.DecisionTreeClassifier(max_depth=2)
clf_low_leaf = tree.DecisionTreeClassifier(min_samples_leaf=2)
clf_medium_leaf = tree.DecisionTreeClassifier(min_samples_leaf=10)
clf_random_splitter = tree.DecisionTreeClassifier(splitter='random')
# The default decision tree has best split strategy, no max depth and 1 sample leaf minimum
clf_default = tree.DecisionTreeClassifier()
trees = [clf_low_depth,clf_low_leaf,clf_medium_leaf,clf_random_splitter,clf_default]
```

Propose an evaluation in terms of training and testing accuracies using $5$-cross validation on two decision trees that have different typologies

```python
from sklearn.model_selection import KFold
# Create 5-cross validation splitter
cv_5 = KFold(n_splits=5)

def createSets(train_index,test_index,train_examples,train_labels):
    #construct Training and Testing data sets
    trainExamples = [elem for index,elem in enumerate(train_examples) if (index in train_index)]
    trainLabels = [elem for index,elem in enumerate(train_labels) if (index in train_index)]
    testExamples = [elem for index,elem in enumerate(train_examples) if (index in test_index)]
    testLabels = [elem for index,elem in enumerate(train_labels) if (index in test_index)]
    return trainExamples,trainLabels,testExamples,testLabels
    
toTrain = [trees[0],trees[3]]
crossScore = [[],[]]
for train_index, test_index in cv_5.split(train_examples):
    trainExamples,trainLabels,testExamples,testLabels = createSets(train_index,test_index,train_examples,train_labels)
    # train it for the index and test it in each iteration
    newFittedTrees = list(map(fit(trainExamples,trainLabels),toTrain))
    
    predictions = list(map(predict(testExamples),newFittedTrees))
    accuracy = [metrics.accuracy_score(pred,testLabels) for pred in predictions]
    crossScore[0].append(accuracy[0])
    crossScore[1].append(accuracy[1])

# Check this under
xAxis = ["1st Fold","2nd Fold","3rd Fold","4th Fold","5th Fold"]
averageAccuracy = [100*sum(elem)/len(elem) for elem in crossScore]
print("First tree average accuracy: ", averageAccuracy[0])
print("Second tree average accuracy: ", averageAccuracy[1])

plt.plot(xAxis,crossScore[0],'ro',label='first tree accuracy')
plt.plot(xAxis,crossScore[1],'yo',label='second tree accuracy')
plt.legend()
plt.show()

# https://towardsdatascience.com/overfitting-vs-underfitting-a-complete-example-d05dd7e19765

# Returns average mean accuracy for the specific model
def accuracy(t,examples,labels):
    results = []
    # Create fold
    for train_index, test_index in cv_5.split(examples):
        trainExamples,trainLabels,testExamples,testLabels = createSets(train_index,test_index,examples,labels)
        fitted = (t.fit(trainExamples,trainLabels))
        prediction = fitted.predict(testExamples)
        results.append(metrics.accuracy_score(prediction,testLabels)*100)
    return mean(results)


# Return cross error list for list of trees
def cross_error(trees,examples,labels):
    accuracies = []
    for t in trees:
        # Compute mean accuracy for each tree
        res = accuracy(t,examples,labels)
        accuracies.append(res)
    return accuracies

toPlot = cross_error(toTrain,train_examples,train_labels)
plt.plot([i for i in range(len(toPlot))],toPlot,'bx')
plt.show()
        
```

Propose an experimental study that shows the transition phase from underfitting to overfitting 

```python
# Create training and testing set
# Progressive learning -> each learning phase is then evaluated and plot the accuracy
def generate_trees(param = "depth"):
    # Check which parameters created more overfitting/ underfitting
    trees = []
    if (param == "depth"):
        for i in range(1,200):
            trees.append(tree.DecisionTreeClassifier(max_depth = i))
    elif (param == "leaf"):
        for i in range(1,200):
            trees.append(tree.DecisionTreeClassifier(min_samples_leaf=i))
    elif (param == "split"):
        for i in range(2,200):
            trees.append(tree.DecisionTreeClassifier(min_samples_split = i))
    return trees

toTrain = generate_trees("depth")
toPlot = cross_error(toTrain,train_examples,train_labels)
plt.plot([i for i in range(len(toPlot))],toPlot,'bx',label="Increasing maximum depth")
plt.legend()
plt.show()

toTrain = generate_trees("leaf")
toPlot = cross_error(toTrain,train_examples,train_labels)
plt.plot([i for i in range(len(toPlot))],toPlot,'bx',label="Increasing minimum sample leaf")
plt.legend()
plt.show()

toTrain = generate_trees("split")
toPlot = cross_error(toTrain,train_examples,train_labels)
plt.plot([i for i in range(len(toPlot))],toPlot,'bx',label="Increasing minimum sample split")
plt.legend()
plt.show()
```
